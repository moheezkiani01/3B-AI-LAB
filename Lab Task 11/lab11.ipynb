```python
import pandas as pd
```

```python
df = pd.read_csv("data.csv")
```

```python
df.head()
```

## Q1. Train Test Splitting

```python
X = df.drop('price', axis=1) 
y = df['price']                
```

```python
from sklearn.model_selection import train_test_split
train_X, test_X, train_y, test_y = train_test_split(X,y, test_size=0.3, shuffle=False) # 80:20, 70:30, 60:40
```

```python
print(train_X.shape, train_y.shape)
print(test_X.shape, test_y.shape)
```

## Q2. Model selection

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
```

```python
model_rfr = RandomForestRegressor()
model_rfr.fit(train_X, train_y)
print(model_rfr)
```

## Q3. Predicting

```python
model_pred_rfr = model_rfr.predict(test_X)
model_pred_rfr
```

## Q4. Accuracy Score

```python
from sklearn.metrics import r2_score, mean_absolute_error

print("MAE:", mean_absolute_error(test_y, model_pred_rfr))
print("R2 Score:", r2_score(test_y, model_pred_rfr))
```
